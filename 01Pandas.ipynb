### Creating and Adding a new Column in DataFrame

```python
# First, let's create our DataFrame
df = pd.DataFrame([('Foreign Cinema', 'Restaurant', 289.0),
                   ('Liho Liho', 'Restaurant', 224.0),
                   ('500 Club', 'bar', 80.5),
                   ('The Square', 'bar', 25.30)],
           columns=('name', 'type', 'AvgBill')
                 )
df
```

```python
# Declaring a new column name with a scalar or list of values
df['Day'] = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']
df
```

```python
## Insert a new column to your DataFrame, Pick a location where you want to add new column
df.insert(loc=3, column="XYZ", value=[2,2,3,4])
df
```

```python
# assign() is a bit like .insert, but you can pass multiple
df.assign(AvgHalfBill=lambda x: x.AvgBill / 2)
print(df)
```

```python
# Passing a Dictonary to your DataFrame
df['Month'] = {'Jan':'Foreign Cinema', 'Feb':'Liho Liho', 'Apr':'500 Club', 'Dec':'Square'}
df
```
